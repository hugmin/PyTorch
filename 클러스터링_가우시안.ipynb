```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
X_train = np.load('C:\DeepLearning\chap11\data\data.npy')
```

```python
gmm = GaussianMixture(n_componenets=2)
gmm.fit(X_train)

print(gmm.means_)
print('\n')
print(gmm.cvariances_)

X, Y = np.meshgrid(np.linspace(-1, 5), np.linspace(-1, 5))
XX = np.array([X.ravel(), Y.ravel()]).T

Z = gmm.score_samples(XX)
Z = Z.reshape((50,50))

plt.contour(X, Y, Z)
plt.scatter(X_train[:0], X_train[:,1])
plt.show()
```
